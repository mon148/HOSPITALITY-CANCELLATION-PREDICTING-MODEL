#### Checking the columns that have null
#### Country

```python
df["country"].unique()
```

```python
df["country"] = df["country"].fillna("unknown")
df["country"].isna().sum()
# df["country"].unique() (This is to confirmed if the null is actually null and has been changed to unknown)
```

### Agent

```python
df["agent"].unique()
```

```python
df["agent"] = df["agent"].fillna(0)
df["agent"].isna().sum()
```

 Filling the null of the agent column with 0 would aid the models not to predict wrongly since 0 indicates nothing was there unlike when filled with mean which would create a wrong assumption in the models training process to believe it actually existed meanwhile it should have been 0

### Company

```python
df["company"].unique()
```

```python
df["company"] = df["company"].fillna(0) # also as that of the 'agent' column
df["company"].isna().sum()
```

```python
df.isna().sum()
```

```python
col_2_drop = ["reservation_status", "arrival_date_year", 
              "agent", "reservation_status_date"
             ]
df.drop(columns=col_2_drop, inplace=True, errors="ignore")
```

Dropping these columns would allow the model to predict base on the right features and not indicating those dropped columns as ranking meanwhile they are just indicators.

```python
df.columns
```

### Removing the 'Hotel' and making only the name exist 

```python
df["hotel"] = df["hotel"].str.replace("Hotel", "").str.strip()
df["hotel"].unique()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

Finally decided to drop the duplicates so as for the model to predict from the direct data imputed and the unique values
